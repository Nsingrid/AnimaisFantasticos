```
import pandas as pd
import matplotlib.pyplot as plt
#scikit learn
from sklearn.cluster import KMeans

df = pd.read_excel('sample_data/pesoaltura.xlsx')
plt.scatter(df['Peso'],df['Altura'])
print(df)
plt.show()
kmeans = KMeans(n_clusters=2)
kmeans.fit(df)
df['cluster'] = kmeans.labels_

centroides = kmeans.cluster_centers_
print(centroides)
plt.scatter(df['Peso'],df['Altura'],c=df['cluster'])
plt.plot(centroides[:,0],centroides[:,1],'rD')
plt.show()
predicao = kmeans.predict([ [20,1], [140,2], [60,1.6]])
print('Clusters dos elementos (20,1), (140,2), (60,1.6): ',predicao)

plt.plot([20,140,60],[1,2,1.6],'k.')
plt.scatter(df['Peso'],df['Altura'],c=df['cluster'])
plt.plot(centroides[:,0],centroides[:,1],'rD')
plt.show()
```
